### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Step 1: Display column names
print("Column Names:")
print(df.columns.tolist())
```

```python
# Step 2: Display data types
print("\nData Types:")
print(df.dtypes)
```

```python
# Step 3: Display summary information
print("\nSummary Information:")
print(df.info())
```

```python
# Step 4: Display basic statistics for numeric columns
print("\nBasic Statistics for Numeric Columns:")
print(df.describe())
```

```python
# Step 5: Display basic statistics for categorical columns
print("\nBasic Statistics for Categorical Columns:")
print(df.describe(include='object'))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

# Step 1: Identify missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)

```

```python
# Step 2: Impute missing values
# Numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_columns:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Step 4: Verify the changes
missing_values_after = df.isnull().sum()
print("\nMissing Values After Handling:")
print(missing_values_after)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Step 1: Calculate value counts for each column
employment_counts = df['Employment'].value_counts()
jobsat_counts = df['JobSat'].value_counts()
yearscodepro_counts = df['YearsCodePro'].value_counts()

# Step 2: Calculate normalized value counts (proportions)
employment_proportions = df['Employment'].value_counts(normalize=True) * 100
jobsat_proportions = df['JobSat'].value_counts(normalize=True) * 100
yearscodepro_proportions = df['YearsCodePro'].value_counts(normalize=True) * 100

# Step 3: Display the results
print("Employment Value Counts:")
print(employment_counts)
print("\nEmployment Proportions (%):")
print(employment_proportions)

print("\nJobSat Value Counts:")
print(jobsat_counts)
print("\nJobSat Proportions (%):")
print(jobsat_proportions)

print("\nYearsCodePro Value Counts:")
print(yearscodepro_counts)
print("\nYearsCodePro Proportions (%):")
print(yearscodepro_proportions)
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

# Step 1: Calculate value counts for JobSat
jobsat_counts = df['JobSat'].value_counts()

# Step 2: Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(jobsat_counts, labels=jobsat_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
plt.title('Distribution of Job Satisfaction Levels')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
from matplotlib_venn import venn2
# Step 1: Extract programming languages
languages_worked_with = set(df['LanguageHaveWorkedWith'].str.split(';').explode().dropna())
languages_want_to_work_with = set(df['LanguageWantToWorkWith'].str.split(';').explode().dropna())

# Step 2: Create a Venn diagram
plt.figure(figsize=(8, 6))
venn2([languages_worked_with, languages_want_to_work_with], set_labels=('Languages Worked With', 'Languages Want to Work With'))
plt.title('Overlap Between Languages Worked With and Languages Want to Work With')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

# Step 1: Group the data by Country and RemoteWork
remote_by_country = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)

# Step 2: Select top N countries for visualization (e.g., top 10)
top_countries = remote_by_country.sum(axis=1).sort_values(ascending=False).head(10).index
remote_by_country_top = remote_by_country.loc[top_countries]

# Step 3: Create a grouped bar chart
remote_by_country_top.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='viridis')
plt.title('Remote Work Preferences by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Remote Work Preference')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
# Step 1: Prepare the data
# Convert JobSat to numeric values
jobsat_mapping = {
    'Very dissatisfied': 1,
    'Slightly dissatisfied': 2,
    'Neutral': 3,
    'Slightly satisfied': 4,
    'Very satisfied': 5
}
df['JobSatNumeric'] = df['JobSat'].map(jobsat_mapping)

# Convert YearsCodePro to numeric (handle non-numeric values)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows with missing values in either column
df_clean = df.dropna(subset=['JobSatNumeric', 'YearsCodePro'])
```

```python
# Step 2: Calculate the Spearman correlation coefficient
from scipy.stats import spearmanr, pearsonr

spearman_corr, spearman_p = spearmanr(df_clean['JobSatNumeric'], df_clean['YearsCodePro'])
print(f"Spearman Correlation Coefficient: {spearman_corr:.2f}")
print(f"P-value: {spearman_p:.4f}")
```

```python
# Step 3: Calculate the Pearson correlation coefficient (optional)
pearson_corr, pearson_p = pearsonr(df_clean['JobSatNumeric'], df_clean['YearsCodePro'])
print(f"Pearson Correlation Coefficient: {pearson_corr:.2f}")
print(f"P-value: {pearson_p:.4f}")
```

```python
# Step 4: Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsCodePro', y='JobSatNumeric', data=df_clean, alpha=0.6, color='blue')
sns.regplot(x='YearsCodePro', y='JobSatNumeric', data=df_clean, scatter=False, color='red')
plt.title('Correlation Between Years of Experience and Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience (YearsCodePro)')
plt.ylabel('Job Satisfaction (JobSatNumeric)')
plt.grid(True)
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

# Step 1: Create a cross-tabulation
cross_tab = pd.crosstab(df['Employment'], df['EdLevel'], normalize='index') * 100
print("Cross-Tabulation of Employment and Education Level (%):")
print(cross_tab)

```

```python
# Step 2: Visualize the results with a stacked bar plot
plt.figure(figsize=(12, 8))
cross_tab.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Education Level by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Percentage of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
